```python
import numpy as np
```

```python
pages = np.arange(440, 2000)
count = 1
lista_personajes = []

for page in pages:
    
    URL = 'https://www.anime-planet.com/characters/all?page=' + str(page)
    r = requests.get(URL)
    soup = BeautifulSoup(r.text, 'lxml')
    personaje_grid = soup.find_all(class_="tableAvatar")
    
    personaje_count = 1 # para el print de seguimiento de descarga
    
    for img in personaje_grid:
        # Print de seguimiento de descarga:
        print('img {} de {}, pag {}/{}'.format(
            personaje_count, len(personaje_grid), page, len(pages)))

        image = img.find('img')['src']

        lista_personajes.append([
            ('image', image)
        ])        
        
        # Pasamos al siguiente id
        count += 1
        personaje_count += 1
```

```python
import pickle

with open("LISTA_DE_URLS_ANIME.txt", "wb") as fp:   #Pickling
    pickle.dump(lista_personajes, fp)

```
